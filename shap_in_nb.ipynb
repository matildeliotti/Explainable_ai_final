```python
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import shap
import matplotlib.pyplot as plt
```

```python
#Loading the dataset
df = pd.read_csv('creditcard.csv')
X = df.drop(['Class'],axis=1)
y = df['Class']
```

```python
#We split into training and testing subsets
train, test, labels_train, labels_test = train_test_split(X, y, test_size=0.2, random_state=7)
print("Training samples:", train.shape[0])
print("Testing samples:", test.shape[0])
```

```python
#Just showing the first lines of the training set
print(train.head())
```

```python
#We train a random forest classifier
rf = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
rf.fit(train, labels_train)
```

```python
#We evaluate the performace in the prediction of the model
y_pred = rf.predict(test)
print("Accuracy: ", accuracy_score(labels_test, y_pred))
```

```python
#We only care about the lines of data with fraudolent transactions, therefore we will isolate them
fraud_list = [row for index, row in test.iterrows() if labels_test[index] == 1]

frauds = pd.DataFrame(fraud_list)
print("Fraudulent transactions found:", len(frauds))
```

```python
#We select one fraudolent transaction to understand why it was classified as such
fraudolent_tran = frauds.iloc[[0]]

#shap explainer for tree models
explainer = shap.TreeExplainer(rf)

#We get the shap values for the selected fraudolent transaction
shap_values = explainer(fraudolent_tran)
```

```python
#waterfall plot to read each feature's contribution to fraudolence detection
import matplotlib.pyplot as plt
shap.plots.waterfall(shap_values[0, :, 1])
plt.show()
```
